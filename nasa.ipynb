#**model training code**

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
import lightgbm as lgb
import time

X_train = pd.read_csv('X_train.csv')
X_test = pd.read_csv('X_test.csv')
y_train = pd.read_csv('y_train.csv').values.ravel()
y_test = pd.read_csv('y_test.csv').values.ravel()

print(f"Train: {X_train.shape}, Test: {X_test.shape}")
print(f"Positive class in train: {y_train.sum()}/{len(y_train)} = {y_train.mean():.2%}")

# ============ MODEL 1: RANDOM FOREST ============
print("\n" + "="*50)
print("TRAINING RANDOM FOREST")
print("="*50)

start = time.time()
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, y_train)
rf_time = time.time() - start

y_pred_rf = rf.predict(X_test)
y_proba_rf = rf.predict_proba(X_test)

print(f"Training time: {rf_time:.1f}s")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf, target_names=['Non-planet','Candidate', 'Planet']))
print(f"AUC: {roc_auc_score(y_test, y_proba_rf, multi_class='ovo', average='weighted'):.4f}")

# ============ MODEL 2: LIGHTGBM ============
print("\n" + "="*50)
print("TRAINING LIGHTGBM")
print("="*50)

start = time.time()
lgbm = lgb.LGBMClassifier(
    n_estimators=200,
    learning_rate=0.1,
    is_unbalance=True,
    random_state=42,
    verbose=-1
)
lgbm.fit(X_train, y_train)
lgbm_time = time.time() - start

y_pred_lgbm = lgbm.predict(X_test)
y_proba_lgbm = lgbm.predict_proba(X_test)

print(f"Training time: {lgbm_time:.1f}s")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_lgbm))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_lgbm, target_names=['Non-planet','Candidate', 'Planet']))
print(f"AUC: {roc_auc_score(y_test, y_proba_lgbm, multi_class='ovo', average='weighted'):.4f}")

# ============ MODEL 3: 1D CNN ============
print("\n" + "="*50)
print("TRAINING 1D CNN")
print("="*50)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

X_train_cnn = np.expand_dims(X_train, axis=2)
X_test_cnn = np.expand_dims(X_test, axis=2)

num_classes = len(np.unique(y_train))
y_train_cnn = to_categorical(y_train, num_classes)
y_test_cnn = to_categorical(y_test, num_classes)

cnn = Sequential([
    Conv1D(32, kernel_size=3, activation='relu', input_shape=(X_train_cnn.shape[1], 1)),
    BatchNormalization(),
    Dropout(0.25),
    MaxPooling1D(pool_size=2),

    Conv1D(64, kernel_size=3, activation='relu'),
    BatchNormalization(),
    Dropout(0.25),
    MaxPooling1D(pool_size=2),

    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.4),
    Dense(num_classes, activation='softmax')
])

cnn.compile(optimizer=Adam(learning_rate=0.0005),
            loss='categorical_crossentropy',
            metrics=['recall'])

start = time.time()
history = cnn.fit(
    X_train_cnn, y_train_cnn,
    validation_split=0.2,
    epochs=30,
    batch_size=256,
    verbose=1
)
cnn_time = time.time() - start

y_pred_cnn = np.argmax(cnn.predict(X_test_cnn), axis=1)
y_proba_cnn = cnn.predict(X_test_cnn)

print(f"\nTraining time: {cnn_time:.1f}s")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_cnn))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_cnn, target_names=['Non-planet','Candidate','Planet']))
print(f"AUC: {roc_auc_score(y_test, y_proba_cnn, multi_class='ovo', average='weighted'):.4f}")

# ============ MODEL 4: STACKING ENSEMBLE ============
print("\n" + "="*50)
print("TRAINING STACKING ENSEMBLE (RF + LGBM)")
print("="*50)

from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

estimators = [
    ('rf', rf),
    ('lgbm', lgbm)
]

stack_model = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42),
    stack_method='predict_proba',
    passthrough=False,
    n_jobs=-1
)

start = time.time()
stack_model.fit(X_train, y_train)
stack_time = time.time() - start

y_pred_stack = stack_model.predict(X_test)
y_proba_stack = stack_model.predict_proba(X_test)

print(f"Training time: {stack_time:.1f}s")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_stack))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_stack, target_names=['Non-planet','Candidate','Planet']))
print(f"AUC: {roc_auc_score(y_test, y_proba_stack, multi_class='ovo', average='weighted'):.4f}")

# ============ COMPARISON ============
print("\n" + "="*50)
print("COMPARISON")
print("="*50)

from sklearn.metrics import recall_score, precision_score, f1_score

results = {
    'Random Forest': {
        'AUC': roc_auc_score(y_test, y_proba_rf, multi_class='ovo', average='weighted'),
        'Recall': recall_score(y_test, y_pred_rf, average='weighted'),
        'Precision': precision_score(y_test, y_pred_rf, average='weighted'),
        'F1': f1_score(y_test, y_pred_rf, average='weighted'),
        'Time': rf_time
    },
    'LightGBM': {
        'AUC': roc_auc_score(y_test, y_proba_lgbm, multi_class='ovo', average='weighted'),
        'Recall': recall_score(y_test, y_pred_lgbm, average='weighted'),
        'Precision': precision_score(y_test, y_pred_lgbm, average='weighted'),
        'F1': f1_score(y_test, y_pred_lgbm, average='weighted'),
        'Time': lgbm_time
    },
    '1D CNN': {
        'AUC': roc_auc_score(y_test, y_proba_cnn, multi_class='ovo', average='weighted'),
        'Recall': recall_score(y_test, y_pred_cnn, average='weighted'),
        'Precision': precision_score(y_test, y_pred_cnn, average='weighted'),
        'F1': f1_score(y_test, y_pred_cnn, average='weighted'),
        'Time': cnn_time
    },
    'Stacking Ensemble RF + LGBM': {
        'AUC': roc_auc_score(y_test, y_proba_stack, multi_class='ovo', average='weighted'),
        'Recall': recall_score(y_test, y_pred_stack, average='weighted'),
        'Precision': precision_score(y_test, y_pred_stack, average='weighted'),
        'F1': f1_score(y_test, y_pred_stack, average='weighted'),
        'Time': stack_time
    }
}

comparison_df = pd.DataFrame(results).T
print(comparison_df.to_string())

best_model = 'Random Forest' if results['Random Forest']['AUC'] > results['LightGBM']['AUC'] else 'LightGBM'
print(f"\nBest model by AUC: {best_model}")

if best_model == 'Random Forest':
    importances = pd.DataFrame({
        'feature': X_train.columns,
        'importance': rf.feature_importances_
    }).sort_values('importance', ascending=False)
else:
    importances = pd.DataFrame({
        'feature': X_train.columns,
        'importance': lgbm.feature_importances_
    }).sort_values('importance', ascending=False)

print("\nTop 15 most important features:")
print(importances.head(15).to_string(index=False))
```
